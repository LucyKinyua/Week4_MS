# Data Preparation

```python
# Checking the dataset for null values;
# 
autolib_df.isnull().sum()
```

```python
# There are a statistically significant number of missing/null values in the two columns,
# That is, 'Displayed comment' and 'Scheduled at'
# Therefore, we could drop them;
# 
autolib_df.drop(['Displayed comment', 'Scheduled at'], axis=1, inplace=True)
autolib_df.head(10)
```

```python
# Checking the shape of the DataFrame to confirm the change has taken effect;
# 
autolib_df.shape
```

```python
# Also, note that some columns are either redundant or irrelevant, e.g.,
# 'Cars' and 'Bluecar counter' always have the same value. Therefore, we could drop 'Cars'
# 'Address', 'ID', 'Geo point', are irrelevant columns
# Therefore, we could drop and leave one of the columns as well;
# 
# 
autolib_df.drop(['Cars',], axis=1, inplace=True)
autolib_df.head(10)
```
